# Data Cleaning

1. Change date rptd, date occ, time occ to datetype

2. Drop Crm Cd 3 & 4, DR_NO

4. Standardize col names

4. Check and drop duplicates

```python
df.info()
```

```python
df['Date Rptd'] = pd.to_datetime(df['Date Rptd'])
```

```python
df['DATE OCC'] = pd.to_datetime(df['DATE OCC'])
```

```python
df['TIME OCC'] = df['TIME OCC'].astype(str).apply(lambda x: x.zfill(4))

df['TIME OCC'] = pd.to_datetime(df['TIME OCC'], format='%H%M').dt.time
```

```python
df.head()
```

```python
df.columns
```

```python
df.drop(columns=['Crm Cd 3','Crm Cd 4','DR_NO'], inplace=True)
```

```python
df.rename(columns={"DATE OCC": "Date Occ",
                   'TIME OCC': 'Time Occ',
                   'AREA':'Area',
                   'AREA NAME': 'Area Name',
                   'Mocodes': 'Mo Codes',
                   'Vict Desc': 'Vict Race',
                   'LOCATION':'Location',
                   'LAT':'Lat',
                   'LON':'Lon'}, inplace=True)
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.to_csv('clean_data/clean_crime_data.csv', index=False)
```
